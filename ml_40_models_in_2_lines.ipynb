```python
path = '/content/Mall_Customers.csv'
```

```python
df = pd.read_csv(path)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:,df.columns != 'Spending Score (1-100)']
y = df['Spending Score (1-100)']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
import lazypredict
```

```python
from lazypredict.Supervised import LazyRegressor
```

```python
from lazypredict.Supervised import LazyClassifier
```

```python
multiple_ML_model = LazyRegressor(verbose=0,ignore_warnings=True,predictions=True)
```

```python
models,predictions = multiple_ML_model.fit(X_train,X_test,y_train,y_test)
```
