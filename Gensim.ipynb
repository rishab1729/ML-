# Gensim 
is a free, open-source library for unsupervised topic modeling and natural language processing in Python. It was designed to handle large volumes of text efficiently, and to extract meaningful structure from unstructured data. Some of the main features of gensim include:

1. Scalability: Gensim is designed to handle large volumes of text, and can efficiently process billions of words.

2. Ease of use: Gensim is implemented in Python and has a user-friendly API, making it easy to use for natural language processing tasks.

3. Quality: Gensim has been developed and tested by a team of experts, and has been widely used in academic research and industry applications.

4. Flexibility: Gensim can be used for a wide range of natural language processing tasks, including topic modeling, document similarity, and text summarization.

Overall, gensim is a powerful tool for working with unstructured text data, and is widely used in natural language processing and machine learning applications.

```python
import gensim.downloader as api

wv = api.load('word2vec-google-news-300')

vec_king = wv['king']
```

```python
vec_king
```

```python
wv.most_similar('man')
```

```python
wv.most_similar('fog')
```

```python
wv.similarity('man','king')
```

```python
import os

from os import path
from wordcloud import WordCloud
```

```python
text = """
Avul Pakir Jainulabdeen Abdul Kalam 15 October 1931 – 27 July 2015) was an Indian aerospace scientist and statesman who served as the 11th President of 
India from 2002 to 2007. He was born and raised in Rameswaram, Tamil Nadu and studied physics and aerospace engineering. He spent the next four decades
as a scientist and science administrator, mainly at the Defence Research and Development Organisation (DRDO) and Indian Space Research Organisation
(ISRO) and was intimately involved in India's civilian space programme and military missile development efforts.He thus came to be known as the Missile 
Man of India for his work on the development of ballistic missile and launch vehicle technology."""
```

```python
wordcloud = WordCloud().generate(text)
```

```python
import matplotlib.pyplot as plt
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
```

# lower max_font_size

```python
wordcloud = WordCloud(max_font_size=40).generate(text)
plt.figure()
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.show()
```
